```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
std = pd.read_csv("D:\Data Science\StudentsPerformance.csv")
```

```python
std.head()
```

```python
std.isna().any()
```

# 1.Males and Females participated in the test

```python
print("No:of Female students:",std['gender'].tolist().count('female'))
print("No:of Male students:",std['gender'].tolist().count('male'))
```

# 2.Students' parental level of education

```python
sns.countplot(x='parental level of education',data=std)
plt.xticks(rotation=45)
plt.xlabel('Level of education',fontsize=10) 
plt.ylabel('Count',fontsize=10)
plt.grid(linestyle="--")
plt.show()
```

Majority of the parents have college education

# 3.a, Average for math, reading and writing based on gender

```python
mean = std.groupby("gender")[["math score","reading score","writing score"]].mean()
```

```python
mean
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=70)
plt.title("Gender vs Maths score")
Gender=std["gender"].value_counts()
y=std.groupby('gender')['math score'].mean()
sns.barplot(Gender.index,y)
plt.grid(linestyle="--")
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=70)
plt.title("Gender vs Reading Score")
Gender=std["gender"].value_counts()
y=std.groupby('gender')['reading score'].mean()
sns.barplot(Gender.index,y)
plt.grid(linestyle="--")
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=70)
plt.title("Gender vs Writing Score")
Gender=std["gender"].value_counts()
y=std.groupby('gender')['writing score'].mean()
sns.barplot(Gender.index,y)
plt.grid(linestyle="--")
```

For maths male students scores the most and for reading and writing scores female students scores the most.
